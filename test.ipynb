```python
# convert jsonl to csv
import pandas as pd

judgment = pd.read_json('fastchat/llm_judge/data/newCombined_bench/model_judgment/gpt-4_single.jsonl', lines=True)
# judgment.to_csv('fastchat/llm_judge/data/newCombined_bench/model_judgment/gpt-4_single.csv')
```

```python
judgment_gpt35 = pd.read_json('fastchat/llm_judge/data/newCombined_bench/model_judgment/gpt-4-1106-preview_single.jsonl', lines=True)
judgment_gpt35
```

```python
# remove the first row
judgment = judgment.iloc[1:]
# judgment.to_csv('fastchat/llm_judge/data/newCombined_bench/model_judgment/gpt-4_single.csv')
judgment.head()
```

```python
original_na = "The assistant's response shows a good level of active listening, empathy, and validation. It acknowledges the user's appreciation and reciprocates it, showing that the assistant values the user's openness and willingness to share. The assistant also reassures the user of its continued support, which contributes to safety and trustworthiness. The response is clear, concise, and encouraging. However, the assistant does not explicitly clarify its role as an AI, which could have contributed to maintaining boundaries and ethical considerations. The response is not judgmental and maintains an open-minded stance. The assistant's response does not have to address a specific concern in this case, so a holistic approach is not applicable here.\n\nActive Listening: [[8]]\nEmpathy & Validation: [[8]]\nSafety & Trustworthiness: [[8]]\nOpen-mindedness & Non-judgment: [[10]]\nClarity & Encouragement: [[9]]\nBoundaries & Ethical: [[7]]\nHolistic Approach: [[N/A]]"
replaced = "The assistant's response shows a good level of active listening, empathy, and validation. It acknowledges the user's appreciation and reciprocates it, showing that the assistant values the user's openness and willingness to share. The assistant also reassures the user of its continued support, which contributes to safety and trustworthiness. The response is clear, concise, and encouraging. However, the assistant does not explicitly clarify its role as an AI, which could have contributed to maintaining boundaries and ethical considerations. The response is not judgmental and maintains an open-minded stance. The assistant's response does not have to address a specific concern in this case, so a holistic approach is not applicable here.\n\nActive Listening: [[8]]\nEmpathy & Validation: [[8]]\nSafety & Trustworthiness: [[8]]\nOpen-mindedness & Non-judgment: [[10]]\nClarity & Encouragement: [[9]]\nBoundaries & Ethical: [[7]]\nHolistic Approach: [[8]]"
replace_dict = {original_na: replaced}
```

```python
judgment.replace(replace_dict, inplace=True)
```

```python
import re
score_categories = ["Active Listening", "Empathy & Validation", "Safety & Trustworthiness", "Open-mindedness & Non-judgment", "Clarity & Encouragement", "Boundaries & Ethical", "Holistic Approach"]

score_dict = {}
for cat in score_categories:
    score_dict[cat] = []
def extract_score(row):
    for cat in score_categories:
        match = re.search(cat + r'.*\[\[(\d+)\]\]', row["judgment"])
        if not match:
            match = re.search(cat + r'.*\n.*\[\[(\d+)\]\]', row["judgment"])
        if match:
            # print(match.group(1))
            score_dict[cat].append(match.group(1))
        else:
            score_dict[cat].append("")
            print(row["question_id"], row["model"])

judgment_gpt35.apply(extract_score, axis=1)
for cat in score_categories:
    judgment_gpt35[cat] = score_dict[cat]
```

```python
# Add a column score to the judgment dataframe that is the average of all the scores, rounded to 2 decimal places
judgment_gpt35["score"] = judgment_gpt35[score_categories].astype(float).mean(axis=1).round(2)

# judgment["score"] = judgment[score_categories].astype(float).mean(axis=1)
judgment_gpt35
```

```python
# get the average score of each model for each prompt
judgment_avg = pd.DataFrame(columns=['model', 'score'])
for model in judgment['model'].unique():
    print(model, judgment[judgment['model'] == model]['score'].mean().round(2))
    # save the average score for each model to one dataframe judgement_avg
    judgment_avg = pd.concat([judgment_avg, pd.DataFrame({'model': [model], 'score': [judgment[judgment['model'] == model]['score'].mean().round(2)]})])
```

```python
judgment_avg.sort_values(by=['score'], ascending=False, inplace=True)
# manually add one column to the dataframe
judgment_avg['version'] = ['finetuned', 'finetuned', 'finetuned', 'base', 'base', 'finetuned', 'base', 'base', 'other']
# judgment_avg.sort_values(by=['score'], ascending=False)
```
